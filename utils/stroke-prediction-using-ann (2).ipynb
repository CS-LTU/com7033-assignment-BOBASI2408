```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = df.drop(['age','gender','work_type','ever_married','Residence_type','smoking_status','stroke'],axis = 'columns')
```

```python
x.head()
```

```python
y = df['stroke']
```

```python
x.head()
```

```python
x.isnull().sum()
```

```python
x['bmi'].fillna(x['bmi'].mean(),inplace=True)
```

```python
# x['bmi'] = x['bmi']/97.600000
```

```python
x.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import normalize,StandardScaler
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
x_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(18,input_shape = (5,),activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(5,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
```

```python
model.fit(x_train , y_train , epochs = 5)
```

```python
model.evaluate(x_test,y_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
arr = model.predict(x_test)
y_pred = []
for val in arr:
    if val>0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
print(classification_report(y_test,y_pred))
```

we have seen above that the positive class samples are very less as compared to the negative class samples. this denotes the dataset is highly imbalanced which results in low precision , recall and f1 score for the postivie class. so we will use SMOTE technique to generate synthetic samples of the positive class

```python
import imblearn
from imblearn.combine import SMOTEENN

sm = SMOTEENN()
x_res, y_res = sm.fit_resample(x,y)
```

```python
y_res.value_counts()
```

```python
x_train , x_test , y_train , y_test = train_test_split(x_res,y_res,test_size=0.3,random_state=42)
```

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
model = keras.Sequential([
    keras.layers.Dense(50,input_shape = (6,),activation='relu'),
    keras.layers.Dense(26,activation='relu'),
    keras.layers.Dense(10,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
```

```python
model.fit(x_train , y_train , epochs =15)
```

```python
model.evaluate(x_test,y_test)
```

```python
arr = model.predict(x_test)
y_pred = []
for val in arr:
    if val>0.5:
        y_pred.append(1)
    else:
        y_pred.append(0)
```

```python
print(classification_report(y_test,y_pred))
```

Now the scores have improves for positive class too. although the accuracy got decreased but this can be due to several reasons like overfitting due to the generated data
